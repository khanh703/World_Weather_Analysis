```python
# Import the random module.
import random

# Import the NumPy module.
# If importing numpy no need to import random - has built-in random module
import numpy as np
```

# random()

```python
random.randint(-90, 90)
```

```python
random.random()
```

### random() returns floating-point decimal between 0 and 1.0
Adding the random floating decimal value to the first random.randint(-90, 89)

```python
random_number = random.randint(-90, 89) + random.random()
random_number
```

## Random latitudes

```python
x = 1
latitudes = []
while x < 11:
    random_lat = random.randint(-90, 89) + random.random()
    latitudes.append(random_lat)
    x += 1
latitudes
```

## Random longitudes

```python
y = 1
longitudes = []
while y < 11:
    random_lon = random.randint(-90,89) + random.random()
    longitudes.append(random_lon)
    y += 1
longitudes
```

# randrange()

Same as random() only require 3rd parameter (step)

```python
random.randrange(-90, 90, step=9)
```

```python
yrr = 1
longitudes = []
while yrr < 11:
    random_lon = random.randrange(-90, 89, step=10) + random.random()
    longitudes.append(random_lon)
    yrr += 1
longitudes
```

## uniform()

Same as random() only generates floating-point decimal between numbers in paranthesis

```python
yu = 1
longitudes = []
while yu < 11:
    random_lon = random.uniform(-90, 89)
    longitudes.append(random_lon)
    yu += 1
longitudes
```

# The NumPy and random Modules

```python
np.random.uniform(-90.12345, 90)
```

```python
# descriptive alternative way
np.random.uniform(low=-90, high=90, size=10)
```

```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500.
latitudes(1500)
```
